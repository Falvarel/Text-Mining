### EJERCICIO 1 Fechas

Vamos a crear expresiones regulares que nos permitan detectar dentro de un texto fechas en distintos formatos. 

```python
import re
```

```python
fechas = ['04-06-23 es la fecha prevista para la entrega','El projecto terminará el 05-09-20',
          '2023-06-04 es la fecha prevista para la entrega','El projecto terminará el 2020-09-05',
          '01-01-21','01-01-2021','01-1-21','01/1/21','12-12-21','12-12-2021','12/12/21','12/12/2021',
          '2023-06-04','2020-09-05','12-30-21','12-25-2021','12-30-21','12-25-2021',
          '2023/06/04','2020/09/05','12/30/21','12/25/2021','12/30/21','12/25/2021']
```

```python
# 1.1 Construye una regex que se active con el siguiente formato dd-mm-yy (e.g. 30-03-20)
```

```python
# 1ª opción:
[x for x in fechas if re.search('^\D*\w{2,2}[-]{1,1}\w{2,2}[-]{1,1}\w{2,2}\D*$',x)]
```

```python
# 2ª opción: si queremos filtrar y que no aparezcan las fechas en formato anglosajón (mm-dd-yy)
[x for x in fechas if re.search('^\D*[0-3]{1}[0-9]{1}[-]{1}[0-1]{1}[0-9]{1}[-]{1}\w{2,2}\D*$',x)]
```

```python
# 1.2 A veces en vez de - se utiliza /. Modifica la regex anterior para que también se active con el siguiente formato:
# dd/mm/yy (e.g. 30/03/20)
# Ojo, debe activarse también con el formato de la celda anterior.
```

```python
# 1ª opción:
[x for x in fechas if re.search('^\D*\w{2,2}[-|/]{1,1}\w{2,2}[-|/]{1,1}\w{2,2}\D*$',x)]
```

```python
# 2ª opción: si queremos filtrar y que no aparezcan las fechas en formato anglosajón (mm-dd-yy)
[x for x in fechas if re.search('^\D*[0-3]{1}[0-9]{1}[-|/]{1}[0-1]{1}[0-9]{1}[-|/]{1}\w{2,2}\D*$',x)]
```

```python
# 1.3 Has llegado hasta aquí pero tú también sabes que no siempre se añade el 0 a los meses que van de enero a septiembre.
# Modifica tu regex para que tenga en cuenta esto. (e.g. 30/3/20)
# Ojo, debe activarse también con el formato de las celdas anteriores.
```

### EJERCICIO 2 Emails 

```python
# Construye una regex que compruebe que un token cumple los requisitos para ser una dirección de email.
```

Para definir la expresión regular solicitada en este ejercicio se han tomado las condiciones necesarias para una dirección de correo electrónico definidas en la Wikipedia. También se han tomado los ejemplos definidos en dicho artículo para comprobar la expresión definida, así como algunos añadidos para comprobar que se cumplen ciertas condiciones específicas.



> Link al artículo:  https://en.wikipedia.org/wiki/Email_address



De acuerdo con el artículo anterior, cualquier dirección de correo electrónico se compone de 2 partes: "Local-part" y "Domain", con el símbolo "@" entre ambas. Los condicionantes principales para ambas partes y que se van a considerar para el presente ejercicio son los siguientes:


#### **Condicionantes:**





---


##### **- Local-part**

Esta parte de la dirección puede ser una cadena entre comillas o no.

###### *CASO 1 (SIN COMILLAS)*

- Pueden emplearse letras minúsculas o mayúsculas de la A a la Z y dígitos del 0 al 9

- Se consideran válidos los siguientes caracteres especiales: !#$%&'*+-/=?^`{|}~

- El caracter "punto" (.) puede emplearse pero nunca como inicio ni final de esta parte y tampoco por duplicado, uno a continuación del siguiente.


###### *CASO 1 (ENTRE COMILLAS)*

- Pueden emplearse letras minúsculas o mayúsculas de la A a la Z y dígitos del 0 al 9

- Se consideran válidos los siguientes caracteres especiales: !#$%&'*+-/=?^`{|}~

- El caracter "punto" (.) puede emplearse pero nunca como inicio ni final de esta parte pero si por duplicado, uno a continuación del siguiente.


Algunos servicios de correo electrónico restringen el uso de algunos de los caracteres especiales definidos con anterioridad. Por ello, dentro de los caracteres especiales válidos sólo vamos a considerar los siguientes: !%+-/_

El tamaño máximo de esta parte de la dirección no puede ser mayor de 64 caracteres.

Por concepto esta parte debería ser sensible al uso de mayúsculas y minúsculas pero en la práctica los servidores suelen considerar correos iguales aquellos similares pero con letras en mayúsculas y/o minúsculas.

---

##### **- Domain**

Esta parte de la dirección estará formada por una serie de "labels" separadas por puntos (sólo 1 punto entre "label" y "label") y limitadas a longitudes de 63 caracteres. Dichas "labels" deberan cumplir las siguientes condiciones:

- Pueden emplearse letras minúsculas o mayúsculas de la A a la Z y dígitos del 0 al 9 (cumpliendo que el "top-level domain" no esté sólo formado por números)

- El caracter "guión medio" (-) puede emplearse pero nunca como inicio ni final de esta parte

---

De acuerdo a las condiciones anteriores, se ha definido una lista de direcciones de correo electrónico, inluyendo direcciones correctas e incorrectas, y la expresión regular para filtrar las primeras.

```python
emails = [
          # EJEMPLOS CORRECTOS:
          'simple@example.com','very.common@example.com','disposable.style.email.with+symbol@example.com','other.email-with-hyphen@example.com',
          'fully-qualified-domain@example.com','user.name+tag+sorting@example.com',
          'x@example.com','example-indeed@strange-example.com','test/test@test.com','admin@mailserver1','example@s.example',
          '" "@example.org','"john..doe"@example.org',  # únicos 2 ejemplos correctos no detectados
          'mailhost!username@example.org','user%example.com@example.org','user-@example.org',

          'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx64@example.com',

          # EJEMPLOS INCORRECTOS:
          'Abc.example.com','A@b@c@example.com','a"b(c)d,e:f;g<h>i[j\k]l@example.com','just"not"right@example.com',
          'this is"not\allowed@example.com','this\ still\"not\\allowed@example.com',
          '123456789012345678901234567890123456789012345678901234567890+xx65@example.com', # único ejemplo incorrectos no detectado
          'i_like_underscore@but_its_not_allowed_in_this_part.example.com','QA[icon]CHOCOLATE[icon]@test.com',
          
          '12345678901234567890123456789012345678901234567890123456789000065@example.com'
          'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx65@example.com',
          '@example.com','..def@example.com','!def@example.com','def..@example.com','Abc..def@example.com','Abc.def@example..com'
          ]
```

```python
[x for x in emails if re.search('^[^\W]([a-zA-Z0-9][/!_%\+\.\-]?){,63}@[a-zA-Z0-9]{,2}[-]?[a-zA-Z]{,63}[\.]?[a-zA-Z]{,63}$',x)]
```

Cabe destacar que se ha intentado definir una expresión regular lo menos compleja posible pero con la capacidad suficiente para filtrar todos los casos de direcciones correctos. Como se puede observar, la expresión regular definida funciona correctamente salvo en 2 casos concretos:

- Uso de comillas ("") en la definición de la "Local-part" (la expresión no es capaz de filtrar estos casos)

      " "@example.org  /  "john..doe"@example.org

- Combinación de caracteres alfanuméricos y especiales en "Local-part" con un nº de caracteres cercanos al límite (la expresión sólo es capaz de filtrar la longitud correcta cuando se utilizan tipos de caracteres similares)

      '123456789012345678901234567890123456789012345678901234567890+xx65@example.com',



A pesar de las limitaciones anteriores, la expresión es capaz de filtrar los casos más generales/habituales al mismo tiempo que se mantiene una cierta sencillez en la propia definición de la expresión, por tanto se considera como válida.

```python
# Definición de función para tokenizar cada palabra y añadir su función en el texto:

def tok_words(texto):
  sent_texto = sent_tokenize(texto)
  tok_sent_words = []

  for frase in sent_texto:
    tokenized_texto = word_tokenize(frase)
    tokenized_words = pos_tag(tokenized_texto)
    tok_sent_words.append(tokenized_words)
  return tok_sent_words
  
```

El análisis de sentimiento se realizará por capítulos, tratando de comprobar si la novela tiene un carácter más "positivo" o "negativo" en general y si hay capítulos más "negativos/positivos" con respecto a otros. Para realizar el análisis de sentimiento léxico (no supervisado) mencionado, se utilizará la librería "TextBlob". Como se verá más adelante, la función de análisis de sentimiento de esta librería nos devuelve dos propiedades: "polarity" y "subjectivity".

- La propiedad "polarity" es un número decimal que se encuentra en el rango de [-1,1], donde 1 corresponde con una frase positiva y -1 una frase negativa. 

- La propiedad "subjectivity" también es un número decimal pero que se encuentra en el rango de [0,1]. Este valor hace referencia a la subjetividad del texto, donde un valor de 0 es muy objetivo y un valor de 1 es muy subjetivo. 

Por tanto podríamos resumir que la propiedad "subjectivity" hace referencia a la opinión personal, la emoción o el juicio, mientras que la propiedad "polarity" se refiere a la información objetiva del texto analizado.


El primer paso para preparar el texto para el análisis va a consistir por tanto en separar la novela estudidada por capítulos.

Posteriormente se filtran las "stopwords" del texto para evitar que tengan mucho peso e influyan de forma determinante en el análisis al ser la palabras más frecuentes de cualquier texto.

```python
# FILTRAMOS "STOPWORDS" EN LOS STRINGS DE CADA CAPÍTULO 
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Def. función para filtrar "stopwords" al string de cada capítulo
def content_text(text):
  stopwords = nltk.corpus.stopwords.words('english')
  new_stopwords = [",",".","'","!",";","?",":","'s","n't","'i"]
  stopwords += new_stopwords
  
  tok_text = word_tokenize(text)
  content = [w for w in tok_text if w.lower() not in stopwords]
  content_text = " ".join(content)
  return content_text
```

```python
LoR_01_chapters_content = []

for chapter in LoR_01_chapters:
  LoR_01_chapters_content.append(content_text(chapter))
```

Acontinuación aplicamos "lemmatización" a cada string del capítulo para tratar de afinar aún más el análisis y que éste pueda recoger las palabras y su sentido de forma más exacta.

```python
# APLICAMOS "LEMMATIZATION" AL STRING DE CADA CAPÍTULO
```

```python
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Def. función para aplicar la "lemmatización" al string de cada capítulo
def lemmatize_text(text):  
  lemmatizer = WordNetLemmatizer() 
  text_words = text.split(" ")
  lemma = []
  #print(text_words)
  for word in text_words:
    lemma.append(lemmatizer.lemmatize(word))
    #print(lemma)
  lemma_text = " ".join(lemma)
  return lemma_text
```

```python
LoR_01_chapters_lemma = []

for chapter in LoR_01_chapters_content:
  LoR_01_chapters_lemma.append(lemmatize_text(chapter))
```

Por último y una vez preparados los strings de cada capítulo procedemos a aplicar el análisis de sentimiento, recogiendo los resultados de las propiedades "polarity" y "subjectivity" para cada uno de ellos.

```python
# APLICAMOS EL MODELO DE ANÁLISIS DE SENTIMIENTO TIPO LÉXICO (NO SUPERVISADO)
```

```python
import textblob
import pandas as pd

# Def. función para aplicar el análisis de sent. al 
# string de cada capítulo y recoger los resultados
def apply_textblob(chapters_string_list):
  chapters_sentiment = []
  i=-1
  
  for chapter_string in chapters_string_list:
    i+=1
    polarity = textblob.TextBlob(chapter_string).sentiment.polarity
    subjectivity = textblob.TextBlob(chapter_string).sentiment.subjectivity
    chapters_sentiment.append(['Chapter '+str(i), polarity, subjectivity])
  
  results_df = pd.DataFrame(chapters_sentiment, columns=['Chapter_Nr','Polarity','Subjectivity'])
  
  return results_df
```

```python
results = apply_textblob(LoR_01_chapters_lemma)
results
```

#### Traductor palabra-palabra (SPA-EN)

```python
hist_spa_01 = ["indio querer hablar vaquero","vaquero disparar","vaquero hombre malo"]
hist_spa_02 = "indio querer hablar vaquero , vaquero disparar , vaquero hombre malo"
hist_spa_03 = 314
```

```python
def spa_en_WordsTrans(text):
  from itertools import chain
  trans_list = []
  i=-1
  print(text)

  # Filtrado y preprocesado de texto a traducir 
  if type(text)==str:
    words_list = text.split(" ")
    #print(words_list)

  if type(text)==list:
    words_list = []
    for x in text:
      x_list = x.split(" ")
      words_list.append(x_list)
    words_list = list(chain.from_iterable(words_list))
    #print(words_list)
  
  if type(text)!=str and type(text)!=list:
    return ("Defined translation function only supports string or list of strings")


  # Traducción del texto procesado
  for word in words_list:
    i+=1
    if wordnet.synsets(word, lang = 'spa') != []:
      en_trans = wordnet.synsets(word, lang = 'spa')[0]
      trans_list.insert(i,en_trans.lemma_names()[0])

    else:
      en_trans = word
      #print(en_trans)
      trans_list.insert(i,en_trans)

  return " ".join(trans_list)
```

```python
spa_en_WordsTrans(hist_spa_02)
```

```python
spa_en_WordsTrans(hist_spa_03)
```
